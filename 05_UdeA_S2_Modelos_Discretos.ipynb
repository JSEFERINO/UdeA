+ _Los modelos probabilsticos discretos se utilizan para describir experimentos aleatorios donde los resultados posibles son un conjunto finito o infinito numerable de valores. Aqu tienes una breve descripcin de algunos modelos discretos y ejemplos asociados a cada uno:_



### 1. **Distribucin Binomial**
La distribucin binomial modela el nmero de xitos en una secuencia de $n$ ensayos independientes de Bernoulli, donde cada ensayo tiene dos posibles resultados: xito o fracaso, y la probabilidad de xito es $p$.

- **Funcin de probabilidad**: $$P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}$$
$$\binom{n}{k} = \frac{n!}{k!(n-k)!}$$
- **Parmetros**:$n$ es el nmero de ensayos y $p$ es la probabilidad de xito en cada ensayo.


### **Distribucin Binomial - Momentos**
- **Valor esperado (Esperanza)**: $$E(X) = n \cdot p$$
- **Varianza**: $$\text{Var}(X) = n \cdot p \cdot (1 - p)$$
- **Desviacin estndar**: $$\sigma(X) = \sqrt{n \cdot p \cdot (1 - p)}$$



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, nbinom, poisson, hypergeom, randint

# @title **Para Graficar cualquier distribucion discreta**
def plot_distribution(dist, values, pmf, title):
    plt.bar(values, pmf)
    plt.title(title)
    plt.xlabel('Valores')
    plt.ylabel('Probabilidad')
    plt.show()
```

```python
# @title **1. Grfica de la Distribucin Binomial con Funcin de probabilidad**: $$P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}$$
n, p = 10, 0.5
x_binom = np.arange(0, n + 1)
pmf_binom = binom.pmf(x_binom, n, p)
plot_distribution(binom, x_binom, pmf_binom, "Distribucin Binomial")
```


#### **Ejemplo 1: - Distribucin Binomial**
+ *Supongamos que lanzamos una moneda justa (probabilidad de xito $p = 0.5$, $n =10$ veces. Cul es la probabilidad de obtener exactamente 6 caras?*

```python
# @title **Solucin ejemplo 1. Binomial -Funcin de probabilidad**: $$P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}$$
from scipy.stats import binom

# Parmetros
n = 10  # Nmero de ensayos
p = 0.5  # Probabilidad de xito
k = 6  # Nmero de xitos

# Probabilidad de obtener exactamente 6 caras
prob_binom_1 = binom.pmf(k, n, p)
print(f"Probabilidad de obtener exactamente 6 caras: P(X = 6) = {prob_binom_1:.5f}")
```

```python
# @title **Momentos Modelo Binomial**
n = 10
p = 0.5
esperanza_binom = n * p
varianza_binom = n * p * (1 - p)
desviacion_binom = np.sqrt(varianza_binom)

print(f"Momentos en el Modelo Binomial")

print(f"Esperanza:E(X) = n*p = ({n})*({p})= {esperanza_binom}")
print(f"Varianza: V(X) =  n*p*(1-p) ={n} * {p} * {(1 - p)} = {varianza_binom}")
print(f"Desviacin estndar: {desviacion_binom:.5f}")
```

### 2. **Distribucin Binomial Negativa**
+ *La distribucin binomial negativa modela el nmero de ensayos requeridos para obtener un nmero fijo de xitos $r$ en ensayos independientes de Bernoulli, donde la probabilidad de xito es $p$.*

- **Funcin de probabilidad**: $$P(X = k) = \binom{k + r - 1}{r - 1} p^r (1 - p)^k$$
- **Parmetros**:$r$ es el nmero de xitos deseados, $p$ es la probabilidad de xito y $k$ es el nmero de fracasos antes de obtener $r$ xitos.

### **Distribucin Binomial Negativa - Momentos**
- **Valor esperado (Esperanza)**: $$E(X) = \frac{r(1 - p)}{p}$$
- **Varianza**: $$\text{Var}(X) = \frac{r(1 - p)}{p^2}$$
- **Desviacin estndar**: $$\sigma(X) = \sqrt{\frac{r(1 - p)}{p^2}}$$



```python
# @title **Grfica de la Distribucin Binomial Negativa** $$P(X = k) = \binom{k + r - 1}{r - 1} p^r (1 - p)^k$$
r, p = 5, 0.7
x_nbinom = np.arange(0, 20)
pmf_nbinom = nbinom.pmf(x_nbinom, r, p)
plot_distribution(nbinom, x_nbinom, pmf_nbinom, "Distribucin Binomial Negativa")
```


#### **Ejemplo 1:- Distribucin Binomial Negativa**
+ *Cul es la probabilidad de obtener 5 xitos antes de 3 fracasos si la probabilidad de xito es $p = 0.7$?*


```python
# @title **2. Solucin Ejemplo 1 -Distribucin Binomial negativa** $$P(X = k) = \binom{k + r - 1}{r - 1} p^r (1 - p)^k$$

from scipy.stats import nbinom

# Parmetros
r = 5  # Nmero de xitos
p = 0.7  # Probabilidad de xito
k = 3  # Nmero de fracasos antes de obtener 5 xitos

# Probabilidad
prob_nbinom_1 = nbinom.pmf(k, r, p)
print(f"Probabilidad de obtener 5 xitos antes de 3 fracasos: P(X=5) = {prob_nbinom_1:.4f}")
```

```python
# @title **Momentos Modelo Binomial Negativo**
r = 5
p = 0.7
esperanza_nbinom = r * (1 - p) / p
varianza_nbinom = r * (1 - p) / (p ** 2)
desviacion_nbinom = np.sqrt(varianza_nbinom)
print(f"Momentos en el Modelo Binomial Negativo")
print(f"Esperanza:E(X) =  {esperanza_nbinom}")
print(f"Varianza: V(X) =  {varianza_nbinom}")
print(f"Desviacin estndar: {desviacion_nbinom}")
```

### 3. **Distribucin de Poisson**
+ *La distribucin de Poisson modela el nmero de eventos que ocurren en un intervalo de tiempo o espacio, bajo la suposicin de que los eventos ocurren con una tasa constante y de forma independiente entre s.*

- **Funcin de probabilidad**: $$P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}$$
- **Parmetros**: $\lambda$ es la tasa promedio de ocurrencia de eventos en el intervalo de tiempo o espacio.


### **Distribucin de Poisson - Momentos**
- **Valor esperado (Esperanza)**: $$E(X) = \lambda$$
- **Varianza**: $$\text{Var}(X) = \lambda$$
- **Desviacin estndar**: $$\sigma(X) = \sqrt{\lambda}$$




```python
# @title **Grfica de la Distribucin de Poisson** $$P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}$$
lambda_poisson = 2
x_poisson = np.arange(0, 10)
pmf_poisson = poisson.pmf(x_poisson, lambda_poisson)
plot_distribution(poisson, x_poisson, pmf_poisson, "Distribucin de Poisson")
```


#### **Ejemplo 1: Distribucin de Poisson**
+ *Una central telefnica recibe en promedio 2 llamadas por minuto. Cul es la probabilidad de que en un minuto se reciban exactamente 3 llamadas?*


```python
# @title  **Solucin Ejemplo 1: Distribucin de Poisson** $$P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}$$
from scipy.stats import poisson

# Parmetros
lambda_poisson = 2  # Tasa promedio de ocurrencia
k = 3  # Nmero de eventos (llamadas)

# Probabilidad
prob_poisson_1 = poisson.pmf(k, lambda_poisson)
print(f"Probabilidad de recibir exactamente 3 llamadas:P(X=3) = {prob_poisson_1:.5f}")
```

```python
# @title  **Momentos en el modelo de Poisson**
lambda_poisson = 2
esperanza_poisson = lambda_poisson
varianza_poisson = lambda_poisson
desviacion_poisson = np.sqrt(varianza_poisson)

print(f"Momentos en el Modelo de Poisson")
print(f"Esperanza:E(X) =  {esperanza_poisson}")
print(f"Varianza: V(X) =  {varianza_poisson}")
print(f"Desviacin estndar: {desviacion_poisson}")
```

### 4. **Distribucin Hipergeomtrica**
+ *La distribucin hipergeomtrica modela el nmero de xitos en una muestra de tamao \(n\) tomada sin reemplazo de una poblacin de tamao \(N\), que contiene \(K\) xitos y \(N - K\) fracasos.*

- **Funcin de probabilidad**: $$P(X = k) = \frac{\binom{K}{k} \binom{N-K}{n-k}}{\binom{N}{n}}$$
- **Parmetros**: \(N\) es el tamao de la poblacin, \(K\) es el nmero de xitos en la poblacin, \(n\) es el tamao de la muestra, y \(k\) es el nmero de xitos en la muestra.

### **Distribucin Hipergeomtrica - Momentos**
- **Valor esperado (Esperanza)**: $$E(X) = n \cdot \frac{K}{N}$$
- **Varianza**: $$\text{Var}(X) = n \cdot \frac{K}{N} \cdot \frac{N - K}{N} \cdot \frac{N - n}{N - 1}$$
- **Desviacin estndar**: $$\sigma(X) = \sqrt{\text{Var}(X)}$$




```python

# @title **Grfica de la Distribucin Hipergeomtrica** $$P(X = k) = \frac{\binom{K}{k} \binom{N-K}{n-k}}{\binom{N}{n}}$$
N, K, n = 30, 10, 5
x_hyper = np.arange(0, n + 1)
pmf_hyper = hypergeom.pmf(x_hyper, N, K, n)
plot_distribution(hypergeom, x_hyper, pmf_hyper, "Distribucin Hipergeomtrica")
```


#### **Ejemplo 1:- Distribucin Hipergeomtrica**
+ *Una urna contiene 10 bolas rojas y 20 bolas azules. Se seleccionan 5 bolas sin reemplazo. Cul es la probabilidad de seleccionar exactamente 3 bolas rojas?*

```python
# @title **Solucin Ejemplo 1:- Distribucin Hipergeomtrica** $$P(X = k) = \frac{\binom{K}{k} \binom{N-K}{n-k}}{\binom{N}{n}}$$
from scipy.stats import hypergeom

# Parmetros
N = 30  # Tamao de la poblacin (total de bolas)
K = 10  # Nmero de xitos en la poblacin (bolas rojas)
n = 5  # Tamao de la muestra
k = 3  # Nmero de xitos en la muestra (bolas rojas)

# Probabilidad
prob_hyper_1 = hypergeom.pmf(k, N, K, n)
print(f"Probabilidad de seleccionar exactamente 3 bolas rojas: P(X=3)={prob_hyper_1:.4f}")
```

```python
# @title  **Momentos en el modelo Hipergeomtrico**
N = 30  # Tamao de la poblacin
K = 10  # xitos en la poblacin
n = 5   # Tamao de la muestra

esperanza_hyper = n * (K / N)
varianza_hyper = n * (K / N) * ((N - K) / N) * ((N - n) / (N - 1))
desviacion_hyper = np.sqrt(varianza_hyper)

print(f"Momentos en el Modelo Hipergeomtrico")
print(f"Esperanza:E(X) =  {esperanza_hyper}")
print(f"Varianza: V(X) =  {varianza_hyper}")
print(f"Desviacin estndar: {desviacion_hyper}")
```

### 5. **Distribucin Uniforme Discreta**
+ *La distribucin uniforme discreta modela una situacin donde todos los valores enteros posibles entre dos lmites $a$ y $b$ tienen la misma probabilidad.*

- **Funcin de probabilidad**: $$P(X = x) = \frac{1}{b - a + 1}$$ para $x = a, a+1, \dots, b$
- **Parmetros**: $a$ es el lmite inferior, $b$ es el lmite superior.

### **Distribucin Uniforme Discreta - Momentos**
- **Valor esperado (Esperanza)**: $$E(X) = \frac{a + b}{2}$$
- **Varianza**: $$\text{Var}(X) = \frac{(b - a + 1)^2 - 1}{12}$$
- **Desviacin estndar**: $$\sigma(X) = \sqrt{\text{Var}(X)}$$


```python
# @title **Grafica de la Uniforme Discreta** $$P(X = x) = \frac{1}{b - a + 1}$$ para $x = a, a+1, \dots, b$
a, b = 1, 6
x_uniform = np.arange(a, b + 1)
pmf_uniform = [1 / (b - a + 1)] * len(x_uniform)
plot_distribution(randint, x_uniform, pmf_uniform, "Distribucin Uniforme Discreta")
```



#### **Ejemplo 1: - Distribucin Uniforme Discreta**
+ *Si un dado se lanza, cul es la probabilidad de obtener un nmero mayor que 4?*


```python
# @title **Solucin ejemplo 1.** $$P(X = x) = \frac{1}{b - a + 1}$$ para $x = a, a+1, \dots, b$
from scipy.stats import randint

# Parmetros
a, b = 1, 6  # Lmite inferior y superior del dado
k = 5  # Nmero mayor que 4

# Probabilidad
prob_uniform_1 = randint.pmf(k, a, b+1)
print(f"Probabilidad de obtener un nmero mayor que 4: P(X>4) = P(X=5)+P(X=6) = 2*{prob_uniform_1:.4f}= {2*prob_uniform_1:.4f}")
```

```python
# @title  **Momentos en el modelo Uniforme discreto**
a, b = 1, 6  # Limites del dado

esperanza_uniform = (a + b) / 2
varianza_uniform = ((b - a + 1) ** 2 - 1) / 12
desviacion_uniform = np.sqrt(varianza_uniform)

print(f"Momentos en el Modelo Uniforme Discreto")
print(f"Esperanza:E(X) =  {esperanza_uniform}")
print(f"Varianza: V(X) =  {varianza_uniform}")
```
